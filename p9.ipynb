```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data = {'x': [1, 2, 3, 4, 5], 'y': [3, 7, 1, 8, 2]}

# Create a line chart
plt.plot(data['x'], data['y'])

# Customize the plot with labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Chart')

# Display the plot
plt.show()
```
